OOP Assignment: 3 Classes with 4 OOP concepts:
- Encapsulation
- Inheritance
- Polymorphism
- Abstraction

Class 1: Product class

```python
class product:                     #encapsulation
    def __init__(self,name,price):
        self.__name=name
        self.__price=price

    def get_name(self):
        return self.__name

    def set_name(self,name):
        self.__name =name

    def get_price(self):
        return self.__price

    def set_price(self,price):
        self.__price=price


    def show_product(self):            #abstraction
        print(f"Product: {self.__name}, Price: {self.__price}")
```

class 2: User class

```python
class user:
    def __init__(self,username,email):
        self.__username=username
        self.__email=email

    def get_username(self):
        return self.__username

    def set_username(self,username):
        self.__username=username

    def get_email(self):
        return self.__email

    def set_email(self, email):
        self.__email=email

    def show_user(self):
        print(f"User: {self.__username}, Email: {self.__email}")
```

class 3: Order class

```python
class order(product):                       # inheritence
    def __init__(self,name,price,quantity):
        super().__init__(name, price)
        self.quantity=quantity

    def show_product(self):               #polymorphism
        total= self.get_price()*self.quantity
        print(f"Product: {self.get_name()}, Price: {self.get_price()}, Quantity: {self.quantity}, Total: {total}")
```

```python
user1=user("abc","abc@123.com")
user1.show_user()

product1=product("book",50)
product1.show_product()

order1=order("book", 50, 4)
order1.show_product()

order1.set_price(80)
order1.show_product()
```
